```python
import numpy as np
import chess

def fen_to_tensor(fen):
    board = chess.Board(fen)
    tensor = np.zeros((12, 8, 8), dtype=np.float32)
    
    piece_to_layer = {
        (chess.PAWN, chess.WHITE): 0,
        (chess.KNIGHT, chess.WHITE): 1,
        (chess.BISHOP, chess.WHITE): 2,
        (chess.ROOK, chess.WHITE): 3,
        (chess.QUEEN, chess.WHITE): 4,
        (chess.KING, chess.WHITE): 5,
        (chess.PAWN, chess.BLACK): 6,
        (chess.KNIGHT, chess.BLACK): 7,
        (chess.BISHOP, chess.BLACK): 8,
        (chess.ROOK, chess.BLACK): 9,
        (chess.QUEEN, chess.BLACK): 10,
        (chess.KING, chess.BLACK): 11,
    }
    
    for square in chess.SQUARES:
        piece = board.piece_at(square)
        if piece:
            layer = piece_to_layer[(piece.piece_type, piece.color)]
            row = square // 8
            col = square % 8
            tensor[layer][row][col] = 1.0
    
    return tensor
```

```python
def move_to_index(move):
    return move.from_square * 64 + move.to_square

def index_to_move(index):
    from_square = index // 64
    to_square = index % 64
    return chess.Move(from_square, to_square)
```

```python
def load_games(pgn_path, your_username):
    X = []
    y = []
    
    pgn_file = open(pgn_path)
    game_count = 0
    
    while True:
        game = chess.pgn.read_game(pgn_file)
        if game is None:
            break
        
        headers = game.headers
        white = headers["White"]
        black = headers["Black"]
        
        # figure out which color you played
        if white == your_username:
            your_color = chess.WHITE
        elif black == your_username:
            your_color = chess.BLACK
        else:
            continue
        
        board = game.board()
        for move in game.mainline_moves():
            # only learn from your moves, not opponent's
            if board.turn == your_color:
                X.append(fen_to_tensor(board.fen()))
                y.append(move_to_index(move))
            board.push(move)
        
        game_count += 1
        if game_count % 100 == 0:
            print(f"Processed {game_count} games...")
    
    print(f"Done. Total games: {game_count}")
    return np.array(X), np.array(y)
```

```python
X, y = load_games("../data/Theni21-white.pgn", "Theni21")
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
X_black, y_black = load_games("../data/Theni21-black.pgn", "Theni21")

# combine both files
X = np.concatenate([X, X_black])
y = np.concatenate([y, y_black])

print("Combined X shape:", X.shape)
print("Combined y shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

# first split off 10% for test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# then split remaining 10% for validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.111, random_state=42)

print("Training:", X_train.shape)
print("Validation:", X_val.shape)
print("Test:", X_test.shape)
```

```python
import torch
import torch.nn as nn

class ChessNet(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.conv_layers = nn.Sequential(
            nn.Conv2d(12, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.ReLU(),
        )
        
        self.fc_layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(256 * 8 * 8, 1024),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(1024, 4096),
        )
    
    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fc_layers(x)
        return x
```

```python
import torch.optim as optim

# setup
model = ChessNet()
optimizer = optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.CrossEntropyLoss()

# convert data to pytorch tensors
X_train_t = torch.tensor(X_train)
y_train_t = torch.tensor(y_train, dtype=torch.long)
X_val_t = torch.tensor(X_val)
y_val_t = torch.tensor(y_val, dtype=torch.long)

print("Setup done")
print("Model parameters:", sum(p.numel() for p in model.parameters()))
```

```python
from torch.utils.data import TensorDataset, DataLoader

# batch the data
train_dataset = TensorDataset(X_train_t, y_train_t)
val_dataset = TensorDataset(X_val_t, y_val_t)

train_loader = DataLoader(train_dataset, batch_size=256, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=256)

print("Batches in training:", len(train_loader))
```

```python
EPOCHS = 20

for epoch in range(EPOCHS):
    # --- training ---
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        predictions = model(X_batch)
        loss = loss_fn(predictions, y_batch)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        correct += (predictions.argmax(1) == y_batch).sum().item()
        total += y_batch.size(0)

    train_acc = correct / total
    avg_loss = total_loss / len(train_loader)

    # --- validation ---
    model.eval()
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            predictions = model(X_batch)
            val_correct += (predictions.argmax(1) == y_batch).sum().item()
            val_total += y_batch.size(0)

    val_acc = val_correct / val_total
    print(f"Epoch {epoch+1}/{EPOCHS} | Loss: {avg_loss:.4f} | Train Acc: {train_acc:.3f} | Val Acc: {val_acc:.3f}")
```

```python
torch.save(model.state_dict(), "../src/model.pth")
print("Model saved")
```

```python
X_test_t = torch.tensor(X_test)
y_test_t = torch.tensor(y_test, dtype=torch.long)

test_dataset = TensorDataset(X_test_t, y_test_t)
test_loader = DataLoader(test_dataset, batch_size=256)

model.eval()
test_correct = 0
test_total = 0

with torch.no_grad():
    for X_batch, y_batch in test_loader:
        predictions = model(X_batch)
        test_correct += (predictions.argmax(1) == y_batch).sum().item()
        test_total += y_batch.size(0)

test_acc = test_correct / test_total
print(f"Test Accuracy: {test_acc:.3f}")
```
